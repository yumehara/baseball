```python
import os
import gc
import numpy as np
import pandas as pd
import lightgbm as lgb
import feather
```

```python
SUBMIT_BALL2_CSV = 'submit/{}/ball2_{}.csv'
SUBMIT_BALL2_F = 'submit/{}/ball2_{}_{}.f'
BALL_2018 = 'intermediate/{}/ball_predict_2018.f'
DIVIDE_TEST = [66265, 132280, 197746, 261104, 327826, 393088, 458426]
divide_num = len(DIVIDE_TEST) + 1
model_No = '30'
```

```python
SUBMIT = SUBMIT_BALL2_CSV.format(model_No, model_No)

# 結果まとめ
result = BALL_2018.format(model_No)
print(result)
df = pd.read_feather(result)
columns = ['predict_straight', 'predict_curve', 'predict_slider', 'predict_shoot', 
           'predict_fork', 'predict_changeup', 'predict_sinker', 'predict_cutball']
```

```python
sample_no = 1
for div in DIVIDE_TEST:
    result = SUBMIT_BALL2_F.format(model_No, model_No, sample_no)
    print(result)
    sample_no = sample_no + 1
    
    temp = pd.read_feather(result)
    for c in columns:
        temp.loc[temp.index <= div, c] = 0
        df[c] = df[c] + temp[c]
```

```python
count = 1
df['count'] = count
for div in DIVIDE_TEST:
    count = count + 1
    df.loc[df.index > div, 'count'] = count
```

```python
for c in columns:
    df[c] = df[c] / df['count']
```

```python
df.drop(columns=['count'], inplace=True)
```
